```python
# 1a. Ensure LapStartDate is datetime and PitStopTime is not NaN
jumbo_df["LapStartDate"] = pd.to_datetime(jumbo_df["LapStartDate"])
```

```python
y_numeric = jumbo_df["PitStopTime"].dt.total_seconds()
```

```python
y_numeric.dtypes
```

----

Create X and y for models.

```python
X = jumbo_df[NUMERICAL_FEATURES + CATEGORICAL_FEATURES].copy()

# Sort all entries based on LapStartDate
X = X.sort_values(by="LapStartDate").reset_index(drop=True)
y = y_numeric.loc[X.index] # Re-align the target (y) with the sorted X
```

```python
X.columns
```

```python
# Define the split point for the data
split_ratio = 0.8
split_index = int(len(X) * split_ratio)
```

```python
# Split Features (X)
X_train_raw = X.iloc[:split_index]
X_test_raw = X.iloc[split_index:]

# Split Target (y)
y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

print(f"Total Pit Stops: {len(X)}")
print(f"Training set size (e.g., 2018-2023): {len(X_train_raw)}")
print(f"Testing set size (e.g., 2024-2025): {len(X_test_raw)}")
print("-" * 30)
```

```python
# --- 3. One-Hot Encoding (Using ONLY Training Data Categories) ---

# CRITICAL STEP: Fit the encoder ONLY on the training data to avoid data leakage.
# Use pd.get_dummies() on both sets, but ensure they align after encoding.

X_train_encoded = pd.get_dummies(X_train_raw, columns=CATEGORICAL_FEATURES, drop_first=True, dtype=int)
X_test_encoded = pd.get_dummies(X_test_raw, columns=CATEGORICAL_FEATURES, drop_first=True, dtype=int)
```

```python
X_train_encoded.columns
```

```python
# Re-alignment Step: This ensures both dataframes have the exact same columns,
# even if a Team/Compound appears in the test set but not the training set (or vice-versa).
# Columns not found in the test set will be added as all zeros (correctly encoding the missing category).

# Get the list of columns from the training set (the authoritative list)
train_cols = X_train_encoded.columns

# Align the test set to the train set columns. Any missing columns in test are filled with 0.
X_test_encoded = X_test_encoded.reindex(columns=train_cols, fill_value=0)
```

```python
# Drop the `LapStartDate` column from the final feature matrix
X_train_final = X_train_encoded.drop(columns=["LapStartDate"])
X_test_final = X_test_encoded.drop(columns=["LapStartDate"])

print(f"Final feature count: {X_train_final.shape[1]}")
print(f"Columns in Train and Test are aligned: {all(X_train_final.columns == X_test_final.columns)}")
print("-" * 30)
```

# --- 4. Model Training (Random Forest Regressor) ---

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Initialize and train a Random Forest Regressor
random_forest_regressor = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
random_forest_regressor.fit(X_train_final, y_train)
```

```python
# Evaluation and Accuracy Check

# Predict on the test set
y_pred = random_forest_regressor.predict(X_test_final)
```

```python
# Calculate metrics (now both y_test and y_pred are float seconds)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Baseline Model Performance on Future Data:")
print(f"Mean Absolute Error (MAE): {mae:.3f} seconds")
print(f"R-squared (R2) Score: {r2:.3f}")
print("-" * 30)

# Insights: 
# The MAE tells you the average prediction error in seconds.
# An MAE of 0.150 seconds, for example, means your model is off by 150 milliseconds on average.
```

# --- 4. Model Training (LinearSVC) ---

```python
import pandas as pd
from sklearn.svm import LinearSVR # Correct model for regression tasks
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, r2_score

linear_svr_pipeline = Pipeline([
    # Step 1: Scaling all features (essential for SVR models)
    ('scaler', StandardScaler()),
    
    # Step 2: Linear Support Vector Regressor
    # C: Regularization parameter (tuning focus)
    # loss: 'epsilon_insensitive' is standard for SVR
    # max_iter: Increased for stable convergence on F1 data
    ('svr', LinearSVR(C=0.1, loss='epsilon_insensitive', max_iter=10000, random_state=42))
])

# --- 2. Training the Model ---

print("Training Linear Support Vector Regressor (SVR)...")
linear_svr_pipeline.fit(X_train_final, y_train)
print("Training complete.")
```

```python
# --- 3. Prediction ---

# Predict on the test set (scaling is applied automatically by the pipeline)
y_pred_svr = linear_svr_pipeline.predict(X_test_final)

# --- 4. Evaluation ---

# Ensure predictions are non-negative (PitStopTime cannot be negative)
y_pred_svr[y_pred_svr < 0] = 0

mae_svr = mean_absolute_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)

print("-" * 45)
print("LinearSVR Model Performance on Future Data:")
print(f"Mean Absolute Error (MAE): {mae_svr:.3f} seconds")
print(f"R-squared (R2) Score: {r2_svr:.3f}")
print("-" * 45)

# Insight: Compare this MAE and R2 against your Random Forest baseline.
# The SVR model is often faster but may require more rigorous tuning of the C parameter.
```
